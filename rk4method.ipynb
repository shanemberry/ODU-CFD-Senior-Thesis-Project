```python
import numpy as np
import math
import plotly.graph_objects as go
```

```python
def dy(x, y):
    return 2*x*y
def f(x):
    return math.exp(x**2)
```

```python
x1 = 0
x2 = 10
h = 0.1
n = int((x2 - x1) / h)
x = 1
y = 1
```

```python
print (x, y, f(x))
for i in range(1, n+1):
    k1 = dy(x, y)
    k2 = dy(x + 0.5*h, y + 0.5*h*k1)
    k3 = dy(x + 0.5*h, y + 0.5*h*k2)
    k4 = dy(x + h, y + h*k3)
    y + h/6*(k1 + 2*k2 + 2*k3 +k4)
    x = x + h
    print (x, y, f(x))
```
